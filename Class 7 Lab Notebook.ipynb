```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
import statsmodels.formula.api as smf

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.pairplot(yelp_data, x_vars=['cool','useful','funny'], y_vars='stars', size=6, aspect=0.7)
```

```python
# create a fitted model
lm1 = smf.ols(formula='stars ~ cool', data=yelp_data).fit() #****left of tilda = response, right of tilda = predictor*****

# print the coefficients
lm1.params
```

```python
# create a fitted model
lm2 = smf.ols(formula='stars ~ useful', data=yelp_data).fit() #****left of tilda = response, right of tilda = predictor*****

# print the coefficients
lm2.params
```
